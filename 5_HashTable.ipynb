## List

```python
stock_prices = []
with open("stock_prices.csv", "r") as f:
    for line in f:
        tokens = line.split(',')
        day = tokens[0]
        price = float(tokens[1])
        stock_prices.append([day, price])
```

```python
stock_prices
```

```python
for element in stock_prices:
    if element[0] == '8-Mar':
        print(element[1])
```

## Dictionary

```python
stock_prices = {}
with open("stock_prices.csv", "r") as f:
    for line in f:
        tokens = line.split(',')
        day = tokens[0]
        price = float(tokens[1])
        stock_prices[day] = price
```

```python
stock_prices
```

```python
stock_prices['9-Mar']
```

### Hash Map


```python
def get_hash(key):
    h = 0
    for char in key:
        h += ord(char)
    return h % 100
```

```python
ord('a')
```

```python
get_hash('a')
```

```python
get_hash('march 28')
```

```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range(self.MAX)]
        
    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    def add(self, key, val):
        h = self.get_hash(key)
        self.arr[h] = val
        
    def get(self, key):
        h = self.get_hash(key)
        return self.arr[h]
    
    def delitem(self, key):
        h = self.get_hash(key)
        self.arr[h] = None
```

```python
t = HashTable()
t.get_hash('march 28')
```

```python
t.add('march 28', 555)
```

```python
t.arr
```

```python
t.get('march 28')
```

```python
t.delitem('march 28')
```
